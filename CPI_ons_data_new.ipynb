```python
df.drop(df.index[452:], inplace=True)
df = df.reset_index(drop=True)
df
```

```python
df_CPIH = df[['Cpih All Items']]

df_CPIH.tail
```

```python
df_CPIH['Cpih All Items'] = df_CPIH.groupby(df_CPIH.index // 3)['Cpih All Items'].mean()
df_CPIH = df_CPIH.dropna()
df_CPIH.reset_index(drop=True)
df_CPIH
```

```python
df_CPIH ['Percentage_Change'] = (df_CPIH['Cpih All Items'].pct_change() * 100).round(1)
df_CPIH['Percentage_Change'].fillna(0, inplace=True)
```
