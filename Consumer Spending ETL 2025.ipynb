I then realised that the additional sheets such TGCN, SDGCN, NDGCN, SERCN and TOURCN has totals such has Total Services,Total Non-Durable Goods, Total Semi-Durable Goods and Total Goods which are irrelevant for the sectors as their collumns which have sub-sectors are already mapped to their relevant sectors, so a loop is created to drop the keys in the dictionary.

```python
for sheet_name in list(All_Data_Sheets.keys()):
    if sheet_name in ['TGCN', 'SDGCN', 'NDGCN', 'SERCN', '0GSCN']: del All_Data_Sheets[sheet_name]

    print(sheet_name)
```

```python
cols_to_drop = []
for sheet_name, df in All_Data_Sheets.items():
    cols_to_drop = [] 
    for col in list(df.columns):
        if df[col].astype(str).str.contains('[x]', case=False, na=False).any():
            cols_to_drop.append(col)
    if cols_to_drop:
        df.drop(columns=cols_to_drop, inplace=True)  

print(f"Sheet '{sheet_name}': Dropped columns with '[x]': {cols_to_drop}")

```

```python
Over_View_Table['Total Actual Expenditure'] = 0.0  # Initialize column

for sheet, sector_dict in subsector_to_sector.items():
    sheet_df = All_Data_Sheets[sheet]  # Get the correct sheet's DataFrame
    for sector, subsector_columns in sector_dict.items():
        try:
            group_sum = sheet_df[subsector_columns].iloc[:, 1:].sum(axis=1)  # Sum subsector columns
            Over_View_Table['Total Actual Expenditure'] += group_sum  # Add to overview
        except KeyError as e:
            print(f"Missing columns in {sheet}: {e}")  # Debug missing columns

print(Over_View_Table['Total Actual Expenditure'])
```

```python
Over_View_Table['Total Actual Expenditure'] = 0.0  # Initialize column

for sheet, sector_dict in subsector_to_sector.items():
    sheet_df = All_Data_Sheets[sheet]  # Get the sheet's DataFrame
    for sector, subsector_columns in sector_dict.items():
        # Only use columns that still exist after dropping
        valid_columns = [col for col in subsector_columns if col in sheet_df.columns]
        if valid_columns:  # Only sum if there are valid columns
            group_sum = sheet_df[valid_columns].iloc[:, 1:].sum(axis=1)  # Sum valid columns, skip first
            Over_View_Table['Total Actual Expenditure'] += group_sum  # Add to overview
        

print(Over_View_Table.columns)
```
