```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
Consumer_spend = pd.read_excel("C:\\Git_test\\Project 2 Docs\\Data Sets\\Consumer_Spend_Transformed.xlsx")
Consumer_spend
```

```python
Totalexp = Consumer_spend[['Time Period', 'Total Expenditure']]

Totalexp.head
```

```python
CPI = pd.read_excel("C:\\Git_test\\Project 2 Docs\\Data Sets\\CPI_ons_data_transformed_CPIH_all.xlsx")
print(CPI.head)
```

```python
Merged_df = pd.concat([Totalexp, CPI], axis=1)

print(Merged_df.columns.tolist)
```

```python
print(list(enumerate(Merged_df.columns)))
```

```python
filtered_df = Merged_df[(Merged_df['Time Period'] >= '2024-01-01') & (Merged_df['Time Period'] <= '2025-01-01')]
```

```python


fig, ax1 = plt.subplots(figsize=(12, 6))

# Primary axis - Total Expenditure (blue line)
ax1.plot(Merged_df ['Time Period'], Merged_df ['Total Expenditure'], color='blue', label='Total Expenditure', linewidth=2)
ax1.set_xlabel('Time Period', fontsize=10)
ax1.set_ylabel('Total Expenditure', fontsize=10, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Secondary axis - Percentage Change (orange line)
ax2 = ax1.twinx()
ax2.plot(Merged_df['Time Period'], Merged_df['Percentage_Change'], color='orange', label='Percentage Change', linewidth=2)
ax2.set_ylabel('Percentage Change (%)', fontsize=10, color='orange')
ax2.tick_params(axis='y', labelcolor='orange')


ax3 = ax2.twinx()
ax3.plot(Merged_df['Time Period'], Merged_df['Cpih All Items'], color='red', label='CPIH', linewidth=2)
ax3.set_ylabel('Total CPIH', fontsize=10, color='red')
ax3.tick_params(axis='y', labelcolor='red')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.tight_layout()
plt.show()
```
