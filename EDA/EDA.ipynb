```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
Consumer_spend = pd.read_excel("C:\\Git_test\\Project 2 Docs\\Data Sets\\Consumer_Spend_Transformed.xlsx")
Consumer_spend
```

```python
Consumer_spend.drop(["Total Actual Expenditure","Total Expenditure","Net Tourism","Total Domestic Expenditure"], axis=1, inplace=True)
```

```python
print(Consumer_spend.columns.tolist)
```

```python
CPI = pd.read_excel("C:\\Git_test\\Project 2 Docs\\Data Sets\\CPI_ons_data_transformed.xlsx")
print(CPI.head)
```

```python
print(CPI.shape)
print(CPI.tail())
```

```python
print(Consumer_spend.shape)
print(Consumer_spend.tail())
```

```python
Merged_df = pd.concat([Consumer_spend, CPI], axis=1)

print(Merged_df.columns.tolist)
```

```python
plot_df = Merged_df.copy()
plot_df = plot_df.rename(columns={
    'Food And Drink': 'Food_And_Drink_Consumer_Spend',
    'Food And Non-Alcoholic Beverages': 'Food_And_Drink_CPI'
})
print(plot_df.columns.tolist())
```

```python
Merged_df.rename(columns={'Miscellaneous Goods And Services':'Miscellaneous', 'Food And Non-Alcoholic Beverages':'Food And Drink','Alcoholic Beverages And Tobacco':'Alcohol, Tobacco And Narcotics','Housing, Water, Elelctricity, Gas And Other Fuels':'Housing','Furnishings, Household Equipment And Routine Maintenance Of The House':'Furniture, Household Equipment And Maintenance','Recreation & Culture':'Recreation And Culture'}, inplace=True)
print(Merged_df.columns.tolist)
```

```python
Consumer_confidence = pd.read_excel("C:\Git_test\Project 2 Docs\Data Sets\CCI_transformed.xlsx")
print(Consumer_confidence.head)
```

```python
print(Consumer_confidence.columns.tolist)
```

```python
Merged_df['CCI Value'] = Consumer_confidence['CCI Value'].values
print(Merged_df.columns.tolist)
```

```python
# Slice one year (2024, quarters 1-4)
year_data = plot_df[plot_df['Time Period'].between('2023-01-01', '2025-01-01')]

plt.figure(figsize=(8, 5))
plt.plot(year_data['Time Period'], year_data['Food_And_Drink_CPI'], label='CPI', color='blue')
plt.ylabel('CPI', color='blue')
plt.twinx()  # Second axis
plt.plot(year_data['Time Period'], year_data['Food_And_Drink_Consumer_Spend'], label='Spending', color='orange')
plt.ylabel('Spending', color='orange')
plt.title('Food/Drink CPI vs Spending (2024)')
plt.legend()
plt.xticks(year_data['Time Period'], 
           ['2023-01-01', '2023-04-01', '2023-07-01', '2023-10-01', '2024-01-01', '2024-04-01', '2024-07-01', '2024-10-01', '2025-01-01'], 
           rotation=45)
plt.tight_layout()
plt.show()
```
