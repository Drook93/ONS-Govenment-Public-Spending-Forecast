```python
df_CPIH = df[['Cpih All Items']]

df_CPIH.tail
```

```python

df_CPIH['Inflation_rate_base_points_monthly'] = df_CPIH['Cpih All Items'] - 100
df_CPIH['Inflation_rate_base_points_monthly'].fillna(0, inplace=True)


df_CPIH['Inflation_rate_monthly'] = (df_CPIH['Cpih All Items'].pct_change() * 100).round(2)
df_CPIH['Inflation_rate_monthly'].fillna(0, inplace=True)
```

```python
df_CPIH
```

```python
df_CPIH
```

```python
df_quarterly = (
    df_CPIH
    .groupby(df_CPIH.index // 3)  # Group every 3 months
    .agg(
        Inflation_rate_quarterly=('Inflation_rate_monthly', 'mean'),
        Inflation_rate_base_points_quarterly=('Inflation_rate_base_points_monthly', 'sum')
    )
    .round(2)
    .reset_index(drop=True)  # Clean index
)
```

```python
df_quarterly 
```

```python
df_quarterly ['Inflation_rate_acceleration_quarterly'] = df_quarterly['Inflation_rate_quarterly'].diff() 
df_quarterly ['Inflation_rate_quarterly'].fillna(0, inplace=True)
```
